## Empirical cumulative distribution function

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd

# generate data
np.random.seed(0)
mu = 200
sigma = 25
x = np.random.normal(mu, sigma, size=100)

# plot the empirical cumulative distribution function
fig, ax = plt.subplots(figsize=(8, 4))
n_bins = 50
n, bins, patches = ax.hist(x, n_bins, normed=1, histtype='step', cumulative=True) 

ax.plot(bins[:-1], n,'r',drawstyle='steps-post')

ax.grid(True)
ax.set_xlim([min(x),max(x)])
ax.set_xlabel('x')
ax.set_ylabel('CDF')

plt.show()
```

## Empirical _complementary_ cumulative distribution function

```python
fig, ax = plt.subplots(figsize=(8, 4))

ax.hist(x, bins=bins, normed=1, histtype='step', cumulative=-1)

ax.grid(True)
ax.set_xlim([min(x),max(x)])
ax.set_xlabel('x')
ax.set_ylabel('Complementary CDF')

plt.show()
```
