## Style sheets

Matplotlib provides several pre-defined style sheets that you may use. See: https://matplotlib.org/users/style_sheets.html. See also the following gallery of style sheets: https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html.

Here is the list of available style sheets:

```python
print(plt.style.available)
```

### Local change of style

You may change the style of plots locally as demonstrated next.

```python
styles = plt.style.available;

for s in styles:
    with plt.style.context((s)):
        plt.figure(figsize=(4,3))
        plt.bar([1,2,3,4], [1,2,3,4], alpha=0.5)
        plt.title(s)
    plt.show()
```

## Matplotlib configuration

You may configure various properties of plots globally to use for plots displayed in your notebook. You may find the following reference useful: https://matplotlib.org/users/customizing.html. Here is a demonstration of this:

```python
# generate random data
np.random.seed(1977)
data = np.random.normal(0, 1, 100).cumsum()

# programatic way using the rc method
inline_rc = dict(plt.rcParams)
inline_rc['lines.linewidth'] = 3 # change the number and see how the graph changes 
plt.rcParams.update(inline_rc) 
#plt.rcParams.update(plt.rcParamsDefault) # uncomment this line if you want to use the default setting

# control the font 
font_options = {'family' : 'monospace',
                'weight' : 'bold',
                'size' : 12} # change the weight and size to see what happens
plt.rc('font', **font_options) # comment out this line to see the difference

#check how to set axes, xtick, ytic, grid, legend
plt.rc('grid', c='0.5', ls='-', lw=1) 
plt.rc('figure', figsize=(5,3))

fig, ax = plt.subplots()
ax.plot(data)
plt.grid(True)
plt.show()
```
